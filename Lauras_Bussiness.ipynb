<h1> Email Campaign Report (2016) </h1>

```python
from __future__ import division
import pandas as pd
import numpy as np
%pylab inline
pylab.rcParams['figure.figsize'] = (12.0, 8.0) # set size of figures"
plt.rcParams.update({'font.size': 22})
```

```python
DF=pd.read_csv('/Users/attiladobi/Downloads/email_campaigns_report.csv')
DF['Time']=pd.to_datetime(DF['Time Sent'])
DF['Count']=(DF['Total Sent']-DF['Total Bounces'])
DF['Opened_Frac']=DF['Total Unique Opens']/DF['Count']*100
DF['Click_Frac']=DF['Total Unique Clicks']/DF['Count']*100
```

```python
cut=DF['Total Sent'].values>50
```

```python
DF['Opened_Frac'][cut].mean(),DF['Click_Frac'][cut].mean()
```

```python
DF['Total Unique Opens'][cut].mean(),DF['Total Unique Clicks'][cut].mean()
```

<h1> Percentage of emails opened and bookings </h1>

```python
plt.plot(DF['Time'][cut],DF['Opened_Frac'][cut],'ob',label='Viewed')
plt.plot(DF['Time'][cut],DF['Click_Frac'][cut],'or',label='Booked')
plt.plot(DF['Time'][cut],np.ones(sum(cut))*DF['Opened_Frac'][cut].mean(),'b')
plt.plot(DF['Time'][cut],np.ones(sum(cut))*DF['Click_Frac'][cut].mean(),'r')
ylabel('email conversion %')
legend(loc='right',numpoints=1)
ax = gca()
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%b\n%Y'))
```

<h1> Email open and click count </h1>

```python
plt.plot(DF['Time'][cut],DF['Total Unique Opens'][cut],'.-b',label='Opened')
plt.plot(DF['Time'][cut],DF['Total Unique Clicks'][cut],'.-r', label='Clicked')
legend(loc='right')
ax = gca()
ax.xaxis.set_major_formatter(matplotlib.dates.DateFormatter('%b\n%Y'))
```

<h1> Increase in customer base </h1>
