```python
#Map
cur.execute("CREATE TABLE emoji_map (id serial PRIMARY KEY,\
    date timestamp,\
    searchTerm text,\
    rank_to integer,\
    search_type text,\
    sort_by text,\
    lang text,\
    timezone text,\
    MAPjson json[],\
    FreqFilter text,\
    FaceFilter text,\
    PatternType text,\
    dateRange timestamp[]\
    );")
conn.commit() #submit change to db
```

```python
word=''
rank_to=20
search_type='emoji' #emoji or skin
if search_type=='skin':
    rank_to=6
sort_by ='lang' #zone or lang
lang='all'
timezone='all'
freq_filter='all'
face_filter='off'
pattern_type='single'
daterange=[datetime.datetime.strptime('2016-03-23 21:51:00','%Y-%m-%d %H:%M:%S'),\
           datetime.datetime.now()]
```

```python
# replace zone with lang when sorting by language
#search all top skin emojis by lang
if face_filter=='off':
    cur.execute("SELECT lang, array_agg(label) as TopEmoji, array_agg(TFreq) as TopCount FROM \
    (SELECT T.lang, T.Label,SUM(T.Freq) as TFreq, rank() OVER (PARTITION BY lang order by SUM(T.Freq) DESC) \
    From (SELECT lang as lang, unnest(emojiSkinLabel) as Label, unnest(emojiSkinCount) as Freq FROM emoji_tweet \
    WHERE (LOWER(text) LIKE '% %')) as T group by T.lang,T.Label) as Q WHERE Q.rank<=30 \
    group by Q.lang;")
result=cur.fetchall()
```

```python
#search top emoji by lang
if face_filter=='off':
    cur.execute("SELECT lang, array_agg(label) as TopEmoji, array_agg(TFreq) as TopCount FROM \
    (SELECT T.lang, T.Label,SUM(T.Freq) as TFreq, rank() OVER (PARTITION BY lang order by SUM(T.Freq) DESC) \
    From (SELECT lang as lang, unnest(emojiLabel) as Label, unnest(emojiCount) as Freq FROM emoji_tweet \
    WHERE (LOWER(text) LIKE '% %')) as T group by T.lang,T.Label) as Q WHERE Q.rank<=30 \
    group by Q.lang;")
result=cur.fetchall()
```

```python
#search all top skin emojis by timezone
if face_filter=='off':
    cur.execute("SELECT zone, array_agg(label) as TopEmoji, array_agg(TFreq) as TopCount FROM \
    (SELECT T.zone, T.Label,SUM(T.Freq) as TFreq, rank() OVER (PARTITION BY zone order by SUM(T.Freq) DESC) \
    From (SELECT time_zone as zone, unnest(emojiSkinLabel) as Label, unnest(emojiSkinCount) as Freq FROM emoji_tweet \
    WHERE (LOWER(text) LIKE '% %')) as T group by T.zone,T.Label) as Q WHERE Q.rank<=30 \
    group by Q.zone;")
result=cur.fetchall()
```

```python
#search all top emojis by timezone
if face_filter=='off':
    cur.execute("SELECT zone, array_agg(label) as TopEmoji, array_agg(TFreq) as TopCount FROM \
    (SELECT T.zone, T.Label,SUM(T.Freq) as TFreq, rank() OVER (PARTITION BY zone order by SUM(T.Freq) DESC) \
    From (SELECT time_zone as zone, unnest(emojiLabel) as Label, unnest(emojiCount) as Freq FROM emoji_tweet \
    WHERE (LOWER(text) LIKE '% %')) as T group by T.zone,T.Label) as Q WHERE Q.rank<=30 \
    group by Q.zone;")
else:
    cur.execute("SELECT zone, array_agg(label) as TopEmoji, array_agg(TFreq) as TopCount FROM \
    (SELECT T.zone, T.Label,SUM(T.Freq) as TFreq, rank() OVER (PARTITION BY zone order by SUM(T.Freq) DESC) \
    From (SELECT time_zone as zone, unnest(emojiLabel) as Label, unnest(emojiCount) as Freq, \
    unnest(emojilabelfacefilter) as FF FROM emoji_tweet WHERE (LOWER(text) LIKE '% %')) as T WHERE (T.FF is TRUE) \
    group by T.zone,T.Label) as Q WHERE Q.rank=1 group by Q.zone;")
result=cur.fetchall()
```

```python
group=[data[0] for data in result]
topEmojis=[data[1][:rank_to] for data in result]
topCount=[map(int,data[2][:rank_to]) for data in result]
topCountSum=[sum(map(int,data[2][:])) for data in result]
topPercent=[["{:0.1f}".format(num) for num in 100*np.array(map(int,data[2][:rank_to]))/sum(map(int,data[2][:]))] for data in result]
```

```python
MAPjson=[json.dumps({'zone':timezone,'emojis':[emj.decode('utf8') for emj in topemojis],'count':count, 'countsum':countsum,\
                     'percent':percent},ensure_ascii=False) for timezone,topemojis,count,countsum,percent in \
                     zip(group,topEmojis,topCount,topCountSum,topPercent)]
```

```python
def convert(lang):
    try:
        val=S[lang].decode('utf8')
    except KeyError:
        val=lang
    return val
```

```python
MAPjson=[{'lang':convert(lang),'emojis':[emj.decode('utf8') for emj in topemojis],'count':count, 'countsum':countsum,\
                     'percent':percent} for lang,topemojis,count,countsum,percent in \
                     zip(group,topEmojis,topCount,topCountSum,topPercent)]
```

```python
MAPjsonDumps=[json.dumps(val,ensure_ascii=False) for val in MAPjson]
```

```python
cur.execute("INSERT INTO emoji_map (\
date,\
searchTerm,\
rank_to,\
search_type,\
sort_by,\
lang,\
timezone,\
MAPjson,\
FreqFilter,\
FaceFilter,\
PatternType,\
dateRange\
)\
VALUES (\
%s,%s,%s,%s,%s,%s,%s,%s::json[],%s,%s,%s,%s\
)",(\
datetime.datetime.utcnow(),\
word,\
rank_to,\
search_type,\
sort_by,\
lang,\
timezone,\
MAPjsonDumps,\
freq_filter,\
face_filter,\
pattern_type,\
daterange,\
))
conn.commit() #submit change to db

```
