```python
ELEVATION_BASE_URL = 'https://maps.googleapis.com/maps/api/elevation/json'

def lookup_elevation(lat, lon):
    url = '%s?locations=%s,%s' % (ELEVATION_BASE_URL, lat, lon)
    result = json.load(urllib.urlopen(url))
    if result['status'] == 'OK':
        return result['results'][0]['elevation']
    else:
        return 'fail'
```

```python
woreda_lat_lon_map = dict()
# RegionName, RegionLat, RegionLon
fpath = '/Users/attiladobi/zenysis/pipeline/auto/ethiopia/bin/shared/data/woreda_mapped.csv'
with open(fpath, 'rb') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    # skip one row
    spamreader.next()
    for row in spamreader:
        woreda_lat_lon_map[row[2]] = {'RegionName': row[0], 'ZoneName': row[1], 'WoredaName': row[2], \
                               'WoredaLat': row[3], 'WoredaLon': row[4], \
                                      'WoredaElevationM': lookup_elevation(row[3], row[4])}
woreda_lat_lon_map
```
