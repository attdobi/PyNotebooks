```python
#for Q1
quarter=1
period=1

quarter_map={1:(1,1),2:(1,2),3:(2,3),4:(2,4),5:(3,1),6:(3,2),7:(4,3),8:(4,4)}
#maps each quarter staring from 4-2016 in terms of (Period,Quarter)
    
def plot_time(t):
    period,quarter = quarter_map[t]

    target_sum=DF.loc[(DF['Quarter']==quarter) & (DF['Period']==period) & (DF['Type'].str.contains('Non-transition'))]\
    .groupby(['Indicator','Subrecipient']).sum().dropna().groupby(level=0).sum()['Value']
    
    toplot=DF.loc[(DF['Quarter']==quarter) & (DF['Period']==period) & (DF['Type'].str.contains('Non-transition'))]\
    .groupby(['Indicator','Subrecipient']).sum().dropna().groupby(level=[1,0]).sum()['Value']/target_sum

    result = pd.DataFrame(toplot).reset_index().pivot(index='Subrecipient', columns='Indicator', values='Value')
    fig=figure()
    fig.set_size_inches(13, 9)
    sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
    sns.heatmap(result, annot=True, fmt=".2f", cmap='viridis',label='big')
    plt.title('Quarter: {:d}, Period: {:d}'.format(quarter,period))
    plt.show()
```

<h1> Pie chart for the breakdown of targets per SR (scroll through each quarter)</h1>

```python
def pie_time(t):
    
    period,quarter = quarter_map[t]

    target_sum=DF.loc[(DF['Quarter']==quarter) & (DF['Period']==period) & (DF['Type'].str.contains('Non-transition'))]\
    .groupby(['Indicator','Subrecipient']).sum().dropna().groupby(level=0).sum()['Value']

    toplot=DF.loc[(DF['Quarter']==quarter) & (DF['Period']==period) & (DF['Type'].str.contains('Non-transition'))]\
    .groupby(['Indicator','Subrecipient']).sum().dropna().groupby(level=[1,0]).sum()['Value']/target_sum
    
    pie_plot=toplot.groupby(level=0).sum()
    
    #Make d3 graph
    
    # Open File to write the D3 Graph
    kwargs1 = {"fontSize": '50'}

    chart = pieChart(width=700, height=500, **kwargs1)
    xdata = pie_plot.index
    #xdata = ['one' , 'two', 'three', 'four']
    ydata1 = pie_plot.values
    extra_serie = {"tooltip": {"y_start": "", "y_end": " cal"}}

    chart.add_serie(title='Q1',name="Serie 1", y=ydata1, x=xdata, extra=extra_serie,**kwargs1)
    chart.buildhtml()

    print('Quarter: {:d}, Period: {:d}'.format(quarter,period))
    display(chart)
```

```python
interact(pie_time, t=(1,8,1) )
```

<h1> Normalize target values to spending </h1>

```python
def quarter_lookup(year,month):
    val=0
    if ((year==2016) & (month>=4) & (month<=6)):
        val=1
    elif ((year==2016) & (month>=7) & (month<=9)):
        val=2
    elif ((year==2016) & (month>=10) & (month<=12)):
        val=3
    else:
        val=0
    return val

# read the entire file into a python array
with open('/Users/attiladobi/Downloads/za_integration.json', 'rb') as f:
    data = f.readlines()

# remove the trailing "\n" from each line
data = map(lambda x: x.rstrip(), data)

# each element of 'data' is an individual JSON object.
# i want to convert it into an *array* of JSON objects
# which, in and of itself, is one large JSON object
# basically... add square brackets to the beginning
# and end, and have all the individual business JSON objects
# separated by a comma
data_json_str = "[" + ','.join(data) + "]"

# now, load it into pandas
DF2 = pd.read_json(data_json_str)

#convert date:
DF2['Date']=pd.to_datetime(DF2['Real_Date'])
DF2['day'] = DF2['Date'].dt.day
DF2['month'] = DF2['Date'].dt.month
DF2['year'] = DF2['Date'].dt.year

#ADD Quarter grouping to the data
DF2['Quarter']=[quarter_lookup(year,month) for year,month in DF2[['year','month']].values]

#Clean up SR names
DF2['subrecipient']=DF2['subrecipient'].replace('nqobile women','nqobile')
#DF2['subrecipient']=DF2['subrecipient'].replace('lifeline nc','lljbh').replace('ll durban','lljbh')\
#.replace('ll fs','lljbh').replace('ll zululand','lljbh')

#standardize names
DF2['field']=DF2['field'].replace('red_umbrella_male_condom_lube','f_1_male_condoms_number_given')
```

```python
spend=DF2.loc[(DF2['field']=='spend') & (DF2['year']<=2016) & (DF2['Quarter']<=2)].groupby(['Quarter','subrecipient']).sum()['val']
#spend
```

<h1> Visual of targets normalized to spending realative to all spending in each SR </h1>

```python
quarter=1
period=1

toplot=DF.loc[(DF['Quarter']==quarter) & (DF['Period']==period) & (DF['Type'].str.contains('Non-transition'))]\
    .groupby(['Indicator','subrecipient']).sum().dropna().groupby(level=[1,0]).sum()['Value']/spend[quarter]

toplot=toplot.dropna()
toplot/=toplot.groupby(level=1).sum()
#toplot

result = pd.DataFrame(toplot).reset_index().pivot(index='subrecipient', columns='Indicator')
fig=figure()
fig.set_size_inches(13, 9)
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sns.heatmap(result, annot=True, fmt=".2f", cmap='viridis',label='big')
plt.title('Quarter: {:d}, Period: {:d}'.format(quarter,period))
plt.show()
```

<h1> Sum of target/spending fraction across each SR for Q1 </h1>

```python
# Open File to write the D3 Graph
pie_plot=toplot.groupby(level=0).sum()

output_file = open('SA_pie_target-to-spend.html', 'w')

kwargs1 = {"fontSize": '50'}

chart = pieChart(name='target-to-spend',width=700, height=500, **kwargs1)
xdata = pie_plot.index
#xdata = ['one' , 'two', 'three', 'four']
ydata1 = pie_plot.values
extra_serie = {"tooltip": {"y_start": "", "y_end": " cal"}}

chart.add_serie(title='Q1',name="Serie 1", y=ydata1, x=xdata, extra=extra_serie,**kwargs1)
chart.buildhtml()

output_file.write(chart.htmlcontent)

# close Html file
output_file.close()

display(chart)
```
