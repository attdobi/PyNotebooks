```python
Events=pd.read_csv('/Users/attiladobi/Downloads/Query_all.csv',index_col='time')
#convert unix time to timestamp and set as primary index
Events['Date'] = pd.to_datetime(Events.index,unit='ms')
Events=Events.set_index('Date')
Events['day'] = pd.to_numeric(Events.index.day)
Events['month'] = pd.to_numeric(Events.index.month)
Events['year'] = pd.to_numeric(Events.index.year)
```

<h1> Find out how many selections people typically choose </h1>

<h3> When grouping by date we can see how many partners people typically select before searching</h3>

<h3> health, target, partners </h3>

```python
names=['health','target','partners']
total_queries=EZ.loc[(EZ['name']=='Run query complete')]['name'].count()
xbins=[1,2,3,4,5]
S=dict()
S['total']=total_queries

for name in names:
    Filter=EZ.columns[EZ.columns.str.contains('selections') & Events.columns.str.contains(name)]
    total_nan=EZ.loc[(EZ['name']=='Run query complete') & (pd.isnull(EZ[Filter[0]]))]['name'].count()
    NumSelect=EZ.loc[(EZ['name']=='Run query complete')][Filter].stack().groupby(level=0).count()
    N,Nbins= np.histogram(NumSelect,xbins)
    ave_per_query=EZ.loc[(EZ['name']=='Run query complete')][Filter].stack().groupby(level=0).count().sum()/total_queries
    S[name]={'NAN':total_nan, 'Nhist':N, 'Nbins':Nbins, 'AvePerQuery':ave_per_query,'NumSelect':NumSelect}
```

<h3> volin plot of number selected per query</h3>

```python
data=[]
aves=[]
for name in names:
    data.append(S[name]['NumSelect'].values) #num select is the raw data before doing hist
    aves.append(mean(S[name]['NumSelect'].values))
    #aves.append(S[name]['AvePerQuery'])
```

```python
sns.set_style("whitegrid")
fig=figure()
ax = sns.violinplot(xlabel=names, data=data,
                     scale="width", palette="Set3",orient='h',gridsize=100,bw=.05)
fig.set_size_inches(10, 7)
sns.set_context("notebook", font_scale=3, rc={"lines.linewidth": 2.5})
plt.xlabel("If Used \n Number Selected Per Query")
plt.yticks(range(3), names)
xlim(0,25)
for y,val in zip(range(3),aves):
    text(val+0.1,y-.1,'{:.2f}'.format(val),fontsize=18)
```

```python
list_of_preFilters=['denominator','endDate','geoType','geography','granularity','relativeDate','startDate']

for name in list_of_preFilters:
    Filter=EZ.columns[EZ.columns.str.contains('selections') & Events.columns.str.contains(name)]
    total_nan=EZ.loc[(EZ['name']=='Run query complete') & (pd.isnull(EZ[Filter[0]]))]['name'].count()
    #breakdown=EZ.loc[(EZ['name']=='Run query complete')].groupby(Filter[0]).count()['name']
    ave_per_query=(total_queries-total_nan)/total_queries
    S[name]={'NAN':total_nan, 'AvePerQuery':ave_per_query} #'BreakDown':breakdown
```

<h1> Plot the most used pre selections </h1>

```python
xdata=names+list_of_preFilters
NotNan=[]
IsNan=[]
for name in xdata:
    IsNan.append(S[name]['NAN'])
    NotNan.append(S['total']-S[name]['NAN'])
notSel=np.array(IsNan) 
Sel=np.array(NotNan)
non_zero=notSel>0
sort_index=notSel[non_zero].argsort()

# Open File to write the D3 Graph
#output_file = open('SA_services.html', 'w')

kwargs1 = {"fontSize": '50'}

chart = multiBarChart(name='preSelect', width=2000, height=400, x_axis_format=None,color_category="category10",**kwargs1)
chart.create_y_axis('yAxis','Times Used in Query')

#xdata = ['one' , 'two', 'three', 'four']
xdata=np.array(xdata)[non_zero][sort_index]
ydata1 = Sel[non_zero][sort_index]
ydata2 = notSel[non_zero][sort_index]

chart.add_serie(name="Selected", y=ydata1, x=xdata, **kwargs1)
chart.add_serie(name="Not Selected", y=ydata2, x=xdata, **kwargs1)

chart.buildhtml()

#output_file.write(chart.htmlcontent)
# close Html file
#output_file.close()

display(chart)
```

<h1> Do the same this but a fractional breakdown </h1>

```python
xdata=names+list_of_preFilters
NotNan=[]
IsNan=[]
for name in xdata:
    IsNan.append(S[name]['NAN'])
    NotNan.append(S['total']-S[name]['NAN'])
notSel=np.array(IsNan) 
Sel=np.array(NotNan)
non_zero=notSel>0
sort_index=notSel[non_zero].argsort()

# Open File to write the D3 Graph
#output_file = open('SA_services.html', 'w')

kwargs1 = {"fontSize": '50'}

chart = multiBarChart(name='preSelectPercent', width=2000, height=400, x_axis_format=None,color_category="category10",**kwargs1)
chart.create_y_axis('yAxis','Percent Used in Query')

#xdata = ['one' , 'two', 'three', 'four']
xdata=np.array(xdata)[non_zero][sort_index]
ydata1 = Sel[non_zero][sort_index]/S['total']*100
ydata2 = notSel[non_zero][sort_index]/S['total']*100

chart.add_serie(name="Selected", y=ydata1, x=xdata, **kwargs1)
chart.add_serie(name="Not Selected", y=ydata2, x=xdata, **kwargs1)

chart.buildhtml()

#output_file.write(chart.htmlcontent)
# close Html file
#output_file.close()

display(chart)
```

```python
def make_pie(breakdown,topn=8):
    pylab.rcParams['figure.figsize'] = (9.0, 9.0) # set size of figures"
    plt.rcParams.update({'font.size': 20})
    figure()
    name=breakdown.axes[0].name.split('.')[-1]
    s_other = pd.Series([breakdown[topn:].sum()], index=['other'])
    breakdown=breakdown[:topn].append(s_other)
    # The slices will be ordered and plotted counter-clockwise.
    labels=[label.split('T0')[0] for label in breakdown.index]
    #labels = [label.split('@')[0] for label in actions_cut.index]
    sizes = breakdown.values
    #colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    #explode = (0.01, 0.01, 0, 0.01)  # only "explode" the 2nd slice (i.e. 'Hogs')
    cmap = plt.cm.Accent
    colors = cmap(np.linspace(0., 1., len(labels)))

    patches, texts, autotexts= plt.pie(sizes, labels=labels, colors=colors,
            autopct='%1.1f%%', pctdistance=0.75, labeldistance=1.05, shadow=True, startangle=90,counterclock=False)
    [text.set_fontsize(20) for text in texts]
    plt.title(name)
    
#make_pie(breakdown,8)
```

<h1> Make with D3JS </h1>

```python
def make_pie_d3(breakdown):
    # Open File to write the D3 Graph
    #output_file = open('SA_pie_target-to-spend.html', 'w')
    name=breakdown.axes[0].name.split('.')[-1]
    kwargs1 = {"fontSize": '100'}

    chart = pieChart(name=name,width=800, height=600,  color_category='category10',**kwargs1)
    xdata = [label.split('T0')[0] for label in breakdown.index]
    #xdata = ['one' , 'two', 'three', 'four']
    ydata1 = breakdown.values
    extra_serie = {"tooltip": {"y_start": "", "y_end": "%"}}

    chart.add_serie(title=name,name="Serie 1", y=ydata1, x=xdata, extra=extra_serie,**kwargs1)
    chart.buildhtml()

    #output_file.write(chart.htmlcontent)
    # close Html file
    #output_file.close()

    display(chart)
#make_pie_d3(breakdown)
```
